```python
# Defining function for plotting graphs

def diff_plots(variable_name):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,15), sharey='all')
    fig.tight_layout(pad=10.0)

    ax1 = sns.countplot(ax=ax1,data=approved_df,x=variable_name)
    ax1.set_title("Refused", fontsize=10)
    ax1.set_ylabel("Number of Loans")
    ax1.set_xticklabels(ax1.get_xticklabels(),rotation=90)

    ax2 = sns.countplot(ax=ax2,data=refused_df,x=variable_name)
    ax2.set_title("Approved", fontsize=10)
    ax2.set_ylabel("Number of Loans")
    ax2.set_xticklabels(ax2.get_xticklabels(),rotation=90)

    ax3 = sns.countplot(ax=ax3,data=canceled_df,x=variable_name)
    ax3.set_title("Canceled", fontsize=10)
    ax3.set_xlabel(variable_name)
    ax3.set_ylabel("Number of Loans")
    ax3.set_xticklabels(ax3.get_xticklabels(),rotation=90)

    ax4 = sns.countplot(ax=ax4,data=unused_df,x=variable_name)
    ax4.set_title("Unused", fontsize=10)
    ax4.set_xlabel(variable_name)
    ax4.set_ylabel("Number of Loans")
    ax4.set_xticklabels(ax4.get_xticklabels(),rotation=90)
    plt.tight_layout()
    plt.show()
```

# Final Conclusion


- Enhance Accessibility for Cash Loans: Given the prevailing preference for cash loans, banks should streamline the process and make cash loans more accessible. Additionally, the equal number of approvals and refusals for revolving loans warrants a review of their desirability.

- Prioritize New Client Acquisition: Focusing on new clients is prudent, as they tend to have higher approval rates. Attracting fresh applicants can bolster the bank’s portfolio.

- Target Female Applicants: Recognizing the larger pool of female candidates, banks should tailor their offerings to meet their needs. Gender-specific strategies can enhance customer engagement.

- Segmentation Based on Education Background: Concentrating efforts on individuals from secondary education backgrounds, who constitute the majority, is strategic. Simultaneously, addressing the needs of higher education        applicants remains essential.

- Strategic Focus on Demographics: While working-class individuals and commercial associates form a significant portion of applicants, attention should also shift toward pensioners and state servants. These emerging segments exhibit higher approval rates and lower default risk.

- Marital Status Considerations: Although married individuals dominate approval statistics, targeting single, separated, and widowed applicants can diversify the bank’s clientele.

- Diversify Loan Types: While cash loans are prevalent, exploring non-cash loans from bank accounts can broaden the range of offerings. Diversification mitigates risk and caters to varying preferences.

- Occupation-Based Risk Assessment: Prioritizing IT and HR staff for loan approvals aligns with their lower default propensity. Customized approaches based on occupation enhance credit quality.

- Shift Away from Urgent Needs: Reevaluating the focus on urgent needs is advisable. Unused offers may correlate with higher default rates, necessitating a strategic shift.

- Selective Housing Loan Approvals: Co-op apartments pose payment challenges; therefore, banks should exercise caution. Prioritizing housing types involving parents or standalone houses/apartments can yield better repayment outcomes.

- Innovative Loan Schemes: Exploring creative loan schemes, especially those benefiting third-party applicants, can address the low credit demand observed for specific purposes (e.g., buying houses, cars). Innovative approaches foster growth and risk management.
